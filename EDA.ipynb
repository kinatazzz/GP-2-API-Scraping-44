# EDA

Собрав полученный датасет, делаем EDA.

```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.drop(columns=['link'], inplace=True)
```

```python
data['Departure'] = pd.to_datetime(data['Departure'], utc=True)
```

```python
data = data[['Airline Title', 'Departure', 'transfers', 'Origin Name', 'Destination Name', 'Duration', 'Price', 'Weather', 'Temperature', 'Places','Title']]
data.head(5)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

### ***Однопеременный анализ***

*Какие самые популярные авиалинии по горящим рейсам:*

```python
data['Airline Title'].nunique()
```

```python
airline_counts = data['Airline Title'].value_counts()
print("Количество рейсов по авиалиниям:")
print(airline_counts)
```

```python
airline_counts.head(5)
```

```python
import plotly.express as px
airline_counts = data['Airline Title'].value_counts()
fig = px.pie(
    airline_counts,
    values=airline_counts.values,
    names=airline_counts.index,
    title='Распределение авиакомпаний',
    color_discrete_sequence=px.colors.qualitative.Pastel
)
fig.update_traces(
    textposition='inside',
    textinfo='percent+label',
    hole=0.3
)

fig.show()
```

*Какая ценовая политика билетов:*

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data['Price'], bins=50, kde=True)
plt.title("Распределение цен на билеты")
plt.xlabel("Цена")
plt.ylabel("Количество предложений")
plt.show()

print(data['Price'].describe())
```

*Насколько долгие перелеты:*

```python
sns.histplot(data['Duration'], bins=50, kde=True)
plt.title("Распределение длительности перелётов")
plt.xlabel("Длительность (мин)")
plt.ylabel("Количество рейсов")
plt.show()
print(data['Duration'].describe())
```

*С каким количеством пересадок больше всего рейсов:*

```python
sns.countplot(x=data['transfers'])
plt.title("Распределение количества пересадок")
plt.xlabel("Количество пересадок")
plt.ylabel("Частота")
plt.show()

print(data['transfers'].value_counts())
```

```python
transfers = data['transfers'].value_counts()

fig = px.pie(
    transfers,
    values=transfers.values,
    names=transfers.index,
    title='Распределение количества пересадок',
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label',
    hole=0.3
)

fig.show()
print(transfers)
```

*Какая температура и погода в пунктах назначения рейсов:*

```python
sns.histplot(data['Temperature'], bins=30, kde=True)
plt.title("Распределение температур в пунктах назначения")
plt.xlabel("Температура (°C)")
plt.ylabel("Частота")
plt.show()
```

```python
weather = data['Weather'].value_counts()

fig = px.pie(
    weather,
    values=weather.values,
    names=weather.index,
    title='Распределение типов погоды',
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label',
    hole=0.3
)

fig.show()
print(weather)
```

*Достопримечательности, места:*

```python
data['Places']
```

```python
data['Places'] = data['Places'].apply(lambda x:x[2:-2].split("', '"))
data['Places'] = data['Places'].apply(lambda x: [place for place in x if 'wikipedia' not in place.lower()])
data['Places']
```

```python
data['place count'] = data['Places'].apply(len)
data['place count']
```

```python
from wordcloud import WordCloud
text_titles = ' '.join(data['Title'].dropna().tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_titles)
plt.figure(figsize=(15, 7.5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Облако слов для описаний предложений Title")
plt.show()
```

### ***Многопеременный анализ***

*Как влияет цена билета в зависимости от длительности полета:*

```python
sns.scatterplot(data=data, x='Duration', y='Price', hue='transfers')
plt.title("Зависимость цены от длительности полета")
plt.xlabel("Длительность (мин)")
plt.ylabel("Цена билета")
plt.show()

print(data[['Price', 'Duration']].corr())
```

```python
fig = px.scatter(
    data,
    x='Duration',
    y='Price',
    color='Duration',
    title='Зависимость цены от длительности перелёта',
    labels={'Duration': 'Длительность (мин)', 'Price': 'Цена билета'},
    hover_data=['Duration'],
    color_continuous_scale=px.colors.sequential.Plasma
)

fig.update_layout(
    xaxis_title='Длительность (мин)',
    yaxis_title='Цена билета',
    showlegend=True
)

fig.show()

print(data[['Price', 'Duration']].corr())
```

```python
import plotly.express as px

fig = px.scatter(
    data,
    x='Temperature',
    y='Price',
    title='Зависимость цены от температуры в пункте назначения',
    labels={'Temperature': 'Температура (°C)', 'Price': 'Цена билета'},
    hover_data=['Origin Name', 'Destination Name'],
    color_discrete_sequence=['blue']
)

fig.update_layout(
    xaxis_title='Температура (°C)',
    yaxis_title='Цена билета',
    showlegend=False
)

fig.show()
```

```python
data['Month'] = data['Departure'].dt.month
data['DayOfWeek'] = data['Departure'].dt.dayofweek

day_names = {0: 'Пн', 1: 'Вт', 2: 'Ср', 3: 'Чт', 4: 'Пт', 5: 'Сб', 6: 'Вс'}
data['DayOfWeek'] = data['DayOfWeek'].map(day_names)

month_names = {1: 'Январь', 2: 'Февраль', 3: 'Март', 4: 'Апрель', 5: 'Май', 6: 'Июнь',
               7: 'Июль', 8: 'Август', 9: 'Сентябрь', 10: 'Октябрь', 11: 'Ноябрь', 12: 'Декабрь'}
data['Month'] = data['Month'].map(month_names)

sns.countplot(x='Month', data=data, order=['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
                                        'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь'])
plt.title("Количество рейсов по месяцам")
plt.xlabel("Месяц")
plt.ylabel("Количество рейсов")
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='DayOfWeek', data=data, order=['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс'])
plt.title("Количество рейсов по дням недели")
plt.xlabel("День недели")
plt.ylabel("Количество рейсов")
plt.show()
```

```python
airline_stats = data.groupby('Airline Title')[['Price', 'Duration']].mean().sort_values('Price')
airline_stats.reset_index(inplace=True)
sns.barplot(x='Price', y='Airline Title', data=airline_stats)
plt.title("Средняя цена билета по авиалиниям")
plt.xlabel("Цена билета")
plt.ylabel("Авиалиния")
plt.show()
```

Анализ маршрутов

```python
route_stats = data.groupby(['Origin Name', 'Destination Name']).agg(
    Count=('Price', 'size'),
    Avg_Price=('Price', 'mean'),
    Avg_Duration=('Duration', 'mean')
).reset_index()

fig = px.scatter(
    route_stats,
    x='Origin Name',
    y='Destination Name',
    size='Count',
    color='Avg_Price',
    hover_data=['Avg_Duration'],
    title='Самые популярные маршруты',
    labels={
        'Origin Name': 'Город отправления',
        'Destination Name': 'Город назначения',
        'Count': 'Количество рейсов',
        'Avg_Price': 'Средняя цена',
        'Avg_Duration': 'Средняя длительность (мин)'
    },
    color_continuous_scale=px.colors.sequential.Plasma
)


fig.update_layout(
    xaxis_title='Город отправления',
    yaxis_title='Город назначения',
    showlegend=True,
    height=600,
    width=1000
)

fig.show()
```

```python
cols = ['Price', 'Duration', 'Temperature', 'transfers']
corr_matrix = data[cols].corr()
print(corr_matrix)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Корреляционная матрица")
plt.show()
```

```python
top_cities = data.groupby('Destination Name')['place count'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette="magma")
plt.title("Топ-10 направлений по количеству достопримечательностей")
plt.xlabel("Среднее количество мест")
plt.ylabel("Город назначения")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
